**6.1** Create a dataframe with the closing prices of 30 different stocks, with 10 from each of the caps.

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# 10 large cap stocks
axisb = pd.read_csv('data/Large_Cap/AXISBANK.csv')
cipla = pd.read_csv('data/Large_Cap/CIPLA.csv')
gail = pd.read_csv('data/Large_Cap/GAIL.csv')
hdfc = pd.read_csv('data/Large_Cap/HDFC.csv')
hdfcb = pd.read_csv('data/Large_Cap/HDFCBANK.csv')
hero = pd.read_csv('data/Large_Cap/HEROMOTOCO.csv')
hindustan = pd.read_csv('data/Large_Cap/HINDUNILVR.csv')
itc = pd.read_csv('data/Large_Cap/ITC.csv')
maruti = pd.read_csv('data/Large_Cap/MARUTI.csv')
titan = pd.read_csv('data/Large_Cap/TITAN.csv')

# 10 mid cap
ajanta = pd.read_csv('data/Mid_Cap/AJANTPHARM.csv')
apollo = pd.read_csv('data/Mid_Cap/APOLLOTYRE.csv')
castrol = pd.read_csv('data/Mid_Cap/CASTROLIND.csv')
dhfl = pd.read_csv('data/Mid_Cap/DHFL.csv')
godrej = pd.read_csv('data/Mid_Cap/GODREJIND.csv')
idbi = pd.read_csv('data/Mid_Cap/IDBI.csv')
jindal = pd.read_csv('data/Mid_Cap/JINDALSTEL.csv')
jubilant = pd.read_csv('data/Mid_Cap/JUBLFOOD.csv')
tatap = pd.read_csv('data/Mid_Cap/TATAPOWER.csv')
voltas = pd.read_csv('data/Mid_Cap/VOLTAS.csv')

# 10 small cap
bajaje = pd.read_csv('data/Small_Cap/BAJAJELEC.csv')
century = pd.read_csv('data/Small_Cap/CENTURYPLY.csv')
fortis = pd.read_csv('data/Small_Cap/FORTIS.csv')
idfc = pd.read_csv('data/Small_Cap/IDFC.csv')
jet = pd.read_csv('data/Small_Cap/JETAIRWAYS.csv')
jk = pd.read_csv('data/Small_Cap/JKLAKSHMI.csv')
lal = pd.read_csv('data/Small_Cap/LALPATHLAB.csv')
lux = pd.read_csv('data/Small_Cap/LUXIND.csv')
pvr = pd.read_csv('data/Small_Cap/PVR.csv')
raymond = pd.read_csv('data/Small_Cap/RAYMOND.csv')


closePrice = pd.DataFrame()   # pronounce it a dataframe
closePrice['axisb'] = axisb.loc[axisb['Series']=='EQ','Close Price'].values
closePrice['cipla'] = cipla.loc[cipla['Series']=='EQ','Close Price'].values
closePrice['gail'] = gail.loc[gail['Series']=='EQ','Close Price'].values
closePrice['hdfc'] = hdfc.loc[hdfc['Series']=='EQ','Close Price'].values
closePrice['hdfcb'] = hdfcb.loc[hdfcb['Series']=='EQ','Close Price'].values
closePrice['hero'] = hero.loc[hero['Series']=='EQ','Close Price'].values
closePrice['hindustan'] = hindustan.loc[hindustan['Series']=='EQ','Close Price'].values
closePrice['itc'] = itc.loc[itc['Series']=='EQ','Close Price'].values
closePrice['maruti'] = maruti.loc[maruti['Series']=='EQ','Close Price'].values
closePrice['titan'] = titan.loc[titan['Series']=='EQ','Close Price'].values
closePrice['ajanta'] = ajanta.loc[ajanta['Series']=='EQ','Close Price'].values
closePrice['apollo'] = apollo.loc[apollo['Series']=='EQ','Close Price'].values
closePrice['castrol'] = castrol.loc[castrol['Series']=='EQ','Close Price'].values
closePrice['dhfl'] = dhfl.loc[dhfl['Series']=='EQ','Close Price'].values
closePrice['godrej'] = godrej.loc[godrej['Series']=='EQ','Close Price'].values
closePrice['idbi'] = idbi.loc[idbi['Series']=='EQ','Close Price'].values
closePrice['jindal'] = jindal.loc[jindal['Series']=='EQ','Close Price'].values
closePrice['jubilant'] = jubilant.loc[jubilant['Series']=='EQ','Close Price'].values
closePrice['tatap'] = tatap.loc[tatap['Series']=='EQ','Close Price'].values
closePrice['voltas'] = voltas.loc[voltas['Series']=='EQ','Close Price'].values
closePrice['bajaje'] = bajaje.loc[bajaje['Series']=='EQ','Close Price'].values
closePrice['century'] = century.loc[century['Series']=='EQ','Close Price'].values
closePrice['fortis'] = fortis.loc[fortis['Series']=='EQ','Close Price'].values
closePrice['idfc'] = idfc.loc[idfc['Series']=='EQ','Close Price'].values
closePrice['jet'] = jet.loc[jet['Series']=='EQ','Close Price'].values
closePrice['jk'] = jk.loc[jk['Series']=='EQ','Close Price'].values
closePrice['lal'] = lal.loc[lal['Series']=='EQ','Close Price'].values
closePrice['lux'] = lux.loc[lux['Series']=='EQ','Close Price'].values
closePrice['pvr'] = pvr.loc[pvr['Series']=='EQ','Close Price'].values
closePrice['raymond'] = raymond.loc[raymond['Series']=='EQ','Close Price'].values

# closePrice[closePrice.isna().values]    # check whether closePrice contains any NaN values
# closePrice.shape    # check the shape of the dataframe  
closePrice.head()   # first few values to understand what the datatframe looks like
```

**6.2** Calculate average annual percentage return and volatility of the stocks.

```
returnPer = closePrice.pct_change().fillna(0)

annualReturn = returnPer.mean()*252
annualVolatility = returnPer.std()*np.sqrt(252)

# printing annual returns and volatility in percentage
print('Returns % : ',annualReturn*100,sep='\n')
print('\n\n---------------------------\n\n')
print('Volatility % : ',annualVolatility*100,sep='\n')

# returnPer.head()    # check the first few values
```

**6.3** Cluster the stocks according to returns and volatility using K-means clustering. Identify optimum number of clusters using the Elbow Curve method.

```
X = np.c_[annualVolatility,annualReturn]    # the data to be clustered
nlist = list(range(1,16))   # trying for a max of 15 clusters
iters = 10    # this might actually not do well for most cases but does ok for this one 
ssd = []

for i in nlist:
  model = KMeans(i,max_iter=iters)
  model.fit(X)
  ssd.append(model.inertia_)

plt.figure(figsize=(10.8,7.2))
plt.plot(nlist,ssd,'-o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances')
plt.show()
```

**6.4** Prepare a separate dataframe to show which stocks belong to same cluster.

```
# the curve is decreasing and didn't flatten out to expose an obvious elbow, but we'll choose 6 clusters as the ssd decreases slowly after that
model = KMeans(5,max_iter=iters)
model.fit(X)

# the dataframe that stores the stocks and the labels they belong to
final = pd.DataFrame({'return':annualReturn,'volatility':annualVolatility,'labels':model.labels_})
centers = model.cluster_centers_

plt.figure(figsize=(14,9))
plt.scatter(final.loc[final['labels']==0,'volatility'],final.loc[final['labels']==0,'return'],c='magenta',label='Cluster 1')
plt.scatter(final.loc[final['labels']==1,'volatility'],final.loc[final['labels']==1,'return'],c='lightgreen',label='Cluster 2')
plt.scatter(final.loc[final['labels']==2,'volatility'],final.loc[final['labels']==2,'return'],c='maroon',label='Cluster 3')
plt.scatter(final.loc[final['labels']==3,'volatility'],final.loc[final['labels']==3,'return'],c='aqua',label='Cluster 4')
plt.scatter(final.loc[final['labels']==4,'volatility'],final.loc[final['labels']==4,'return'],c='orange',label='Cluster 5')
plt.scatter(centers[:,0],centers[:,1],marker='*',c='indigo',s=200,label='Cluster Centers')
for i in final.index:
  plt.annotate(i,(final.loc[i,'volatility'],final.loc[i,'return']))
plt.legend()
plt.show()
```
